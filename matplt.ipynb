## zorder and alpha
#### zorder makes which line to appear first and alpha makes thinner or thicker of the line

```python
plt.plot([2,4,6],[11,21,42],linestyle='-',marker='.',markersize=15,linewidth=12,markeredgecolor='k',alpha=0.65,zorder=2)#makes thinner
plt.plot([2,4,6],[11,40,42],linestyle='-',marker='.',markersize=15,linewidth=12,markeredgecolor='k',zorder=1)
```

## xticks and yticks with xlim and ylim

```python
import numpy as n
plt.plot([3,5,8],[43,45,56],linestyle='--',marker='^',markersize=14,markerfacecolor='black',linewidth=4)
plt.xlabel('x ax',size=23,labelpad=5,color='yellow',backgroundcolor='red')
plt.ylabel('yaxis',size=33,labelpad=18,color='black',backgroundcolor='green')
plt.title('subbu graph',size=13,color='black',pad=42,loc='left',backgroundcolor='yellow')
plt.title('subbu',size=23,color='green',pad=32,loc='right',backgroundcolor='pink')
plt.grid()
plt.xticks([3,5,8],['s','r','p'],size=21,color='green',rotation='vertical',backgroundcolor='red')
plt.yticks(n.arange(40,59,3),['s','u','b','b','u','t','s'],size=23,color='red',backgroundcolor='yellow')
plt.show()
```

## sin and cos graphs

```python
import numpy as n   # must and should match the xticks with previous numbers
a=[0,30,45,60,90]
b=[i*(n.pi/180) for i in a]
print(b)                    # to convert into radiance
plt.plot(a,n.sin(b),marker='+',markeredgecolor='green',color='pink')# here a is 0,30,45... and b is sin(0),sin(30)....
plt.plot(a,n.cos(b),marker='^',color='red',markeredgecolor='black')
plt.xticks(a,['sin0','sin30','sin45','sin60','sin90'],size=12,color='red',backgroundcolor='yellow')
plt.yticks([0.0,0.4,0.6,0.8,1.0],['cos0','cos30','cos45','cos60','cos90'],size=11,color='black',backgroundcolor='white')
plt.show()
```

```python
t=n.arange(0,4*n.pi,0.5)
print(t)
plt.plot(t,n.sin(t),marker='+',label='sin',linestyle='--')
plt.plot(t,n.cos(t),marker='^',label='cos',linestyle=':')
plt.legend() # label is the sin and cos for legend
plt.show()
```

## legend function

```python
t=n.arange(0,4*n.pi,0.5)
print(t)
plt.plot(t,n.sin(t),marker='+',label='sin',linestyle='--')
plt.plot(t,n.cos(t),marker="^",label='cos',linestyle=':')
plt.legend(['sinc','cose'],loc=(0,1),ncol=2,title='my world',facecolor='green',title_fontsize=20)
# ncol prints no of lines col prints where is location
plt.show()
```

###  bar graphs

```python
a=['sun','moon','star','plan']
b=[12,22,32,43]
wid=[0.2,0.4,0.6,0.8]
bot=[2,10,20,30]
col=['red','orange','green','blue']
plt.bar(a,height=b,width=wid,bottom=bot,color=col,align='edge')#name,heig,wid,bottom,align is used for see sun to start frm middle
plt.show()
```
